### Ride Regression ###

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_s = scaler.transform(x_train)
x_test_s = scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)
lm_r.fit(x_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(x_test_s))
```

Validation curve to get diff alpha(/\) values. So to asses best r2 score for alpha

```python
from sklearn.model_selection import validation_curve 
```

below logspace will generate 100 values btw 10^-2 to 10^-8 

```python
param_range = np.logspace(-2,8,100)
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), x_train_s, y_train, "alpha", param_range, scoring='r2')
```

```python
print(train_scores)
```

```python
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis=1)
```

```python
test_mean = np.mean(test_scores, axis=1)
```

```python
test_mean
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range), y=test_mean)
```

```python
np.where(test_mean == max(test_mean))
```

```python
param_range[34]
```

```python
lm_r_best = Ridge(alpha = param_range[34])
```

```python
lm_r_best.fit(x_train_s, y_train)
```

```python
r2_score(y_train, lm_r_best.predict(x_train_s))
```

```python
r2_score(y_test, lm_r_best.predict(x_test_s))
```

```python
mean_squared_error(y_train, lm_r_best.predict(x_train_s))
```

```python
mean_squared_error(y_test, lm_r_best.predict(x_test_s))
```
